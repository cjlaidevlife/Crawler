```python
# import package
import pandas as pd
import numpy as np
```

```python
url =('https://raw.github.com/pandas-dev/pandas/'
      'master/pandas/tests/data/tips.csv')
df = pd.read_csv(url)
print('資料共有：',df.shape[0],'筆，特徵共有：',df.shape[1],'個')
df.head()
```

```python
# 原本:
#df[['total_bill','tip']].head()
# 為了方便:
colname = df.columns
df[[colname[0],colname[1]]].head()
# like SQL 
# select `total_bill`,`tip`
# from df
# limit 5;
```

```python
df.assign(new_column=df[colname[1]]/df[colname[0]]).head()
# like SQL 
# select *,`tip`/`total_bill` as `new_column`
# from df
# limit 5;
```

```python
df[df[colname[5]]=='Dinner'].head()
# like SQL
# select *
# from df
# where `time`='Dinner'
# limit 5;
```

```python
x = df[df[colname[5]]=='Dinner']
print('早餐時間抽菸的比例：')
x[colname[3]].value_counts()
like SQL
# select `smoker`,count(`smoker`)
# from df
# where `time`='Dinner'
# group by `smoker`;
```

```python
df[(df[colname[2]]=='Male') & (df[colname[3]]=='No')]
# like SQL 
# select *
# from df
# where (`sex`='Male') and (`smoker`='No');
```

```python
df[(df[colname[2]]=='Male') | (df[colname[3]]=='No')]
# like SQL 
# select *
# from df
# where (`sex`='Male') or (`smoker`='No');
```
